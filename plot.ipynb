```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
filename1 = 'log1_2columns.csv'
filename2 = 'log2_2columns_target_2500.csv'
filename3 = 'log3_2columns_target_5000.csv'
df1 = pd.read_csv(filename1)
df1 = df1.drop(df1.columns[0], axis = 1)
df2 = pd.read_csv(filename2)
df2 = df2.drop(df2.columns[0], axis = 1)
df3 = pd.read_csv(filename3)
df3 = df3.drop(df3.columns[0], axis = 1)
```

```python
x = df1.iloc[:,0]
y1 = df1.iloc[:,1]
y2 = df2.iloc[:,1]
y3 = df3.iloc[:,1]
#x.tail()
```

```python
fig = plt.figure()
plt.plot(x, y1, label = 'No target')
plt.plot(x, y2, label = 'Target = 2500')
plt.plot(x, y3, label = 'Target = 5000')
plt.legend(loc = 'upper left')
plt.xlabel("Number of steps", fontsize = 14)
#plt.ylabel('Scores', fontsize = 14, fontweight = 'bold')
plt.ylabel("Scores", fontsize = 14)
```

```python
fig.savefig('3 target sizes.png', dpi = 600)
```
