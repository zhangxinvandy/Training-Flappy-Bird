```python
import numpy as np
import pandas as pd 
import matplotlib as plt 
```

```python
filename1 = 'log.txt'
filename2 = 'log_target_2500.txt'
filename3 = 'log_target_5000.txt'
df1 = pd.read_csv(filename1, delim_whitespace = True)

df1 = pd.DataFrame(data = df1)
df1_2cols = df1.iloc[:, [5, 7]]
df1_2cols = df1_2cols.iloc[0:330000]
df1_2cols.columns = ['Step', 'Score']
df1_2cols.tail()
```
